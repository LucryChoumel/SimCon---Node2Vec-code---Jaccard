```python
def name_similarity(instance1, instance2):
    # Obtenez les noms des instances (vous devrez peut-être ajuster cela en fonction de votre ontologie)
    name1 = str(instance1)
    name2 = str(instance2)

    # Calculez la similarité de Levenshtein entre les noms des instances
    similarity_ratio = fuzz.ratio(name1, name2) / 100.0  # Normalisé entre 0 et 1

    return similarity_ratio


def getConcepts(ontoFile):
    """
    This function uses a taxonomy to create two lists with a one to one correspondance
    between child and parent
    """
    pko_onto = Graph()
    pko_onto.parse(ontoFile)
    children = []
    parent = []
    for sub, obj in pko_onto.subject_objects(predicate=RDFS.subClassOf):
        children.append(sub.rsplit('#')[-1])
        parent.append(obj.rsplit('#')[-1])
    return parent, children


def createChains(parent, children):
    """
    This function uses the two list on children and parents to create
    list of all lineages of the corresponding taxonomy
    """
    chains = []
    i = 0
    while i < len(children):
        chain = []
        chain.append(children[i])
        chain.append(parent[i])
        j = i
        while parent[j] in children:
            pos = children.index(parent[j])
            chain.append(parent[pos])
            j = pos
        # print(i+1, "---", chain)
        # print("###", chains)
        if chains:
            for element in chains:
                flag = 0
                if len(element) > len(chain):
                    # print(element, "++", chain)
                    for idx in range(len(element) - len(chain) + 1):
                        if element[idx: idx + len(chain)] == chain:
                            flag = 1
                            break
                elif len(element) < len(chain):
                    # print(element, "--", chain)
                    for idx in range(len(chain) - len(element) + 1):
                        if chain[idx: idx + len(element)] == element:
                            flag = 2
                            break
                if flag != 0:
                    break
            if flag == 2:
                pos = chains.index(element)
                chains[pos] = chain
            elif flag == 0:
                chains.append(chain)
        else:
            chains.append(chain)
        i += 1
    return chains


def consim(filettl, concept1, concept2):
    """
    This function select lineage for each concept, find their common ancestor
    and calculate their similarity
    """
    parent, children = getConcepts(filettl)
    # print("parent", parent)
    # print("\n\n")
    # print("children", children)
    # print("\n\n")
    chains = createChains(parent, children)
    # print("\n\n")
    # print("chains", chains)
    # print("\n\n")
    if concept1 and concept2:
        pos1 = -1
        pos2 = -1
        i = 0
        while i < len(chains):
            if any(concept1 in concept for concept in chains[i]):
                pos1 = i
                break
            i += 1
##        if i < len(chains):
##            print(concept1, "--", chains[pos1])
        c = chains[pos1]
        chain1 = c[c.index(concept1):]
        # print(concept1, "--", chain1)
        i = 0
        while i < len(chains):
            if any(concept2 in concept for concept in chains[i]):
                pos2 = i
                break
            i += 1
##        if i < len(chains):
##            print(concept2, "--", chains[pos2])
        c = chains[pos2]
        chain2 = c[c.index(concept2):]
        # print(concept2, "--", chain2)
        intersect = list(set(chain1).intersection(set(chain2)))
        # print("intersect", intersect)
        nr2 = len(intersect)*2
        n = len(chain1)-len(intersect)+len(chain2)-len(intersect)+len(intersect)*2
##        print("concept similarity between:", concept1, "--", concept2, "= ", nr2/n)
        return nr2/n

#if __name__ == "__main__":
 #   consim("pkmoontology.ttl", "Event", "Event")
    

# Demandez à l'utilisateur de saisir le nom d'un nœud
ressource1 = input("Entrez le nom du nœud : ")
ressource2 = input("Entrez le nom du nœud : ")
bonneRelation = input("Entrez la relation: ")

# Recherchez le numéro du nœud dans le graphe
node_number1 = None
node_number2 = None
node_name1 = None
node_name2 = None
node_name3=None
concept1=None
concept2=None

for node in Graph1.nodes():
    if Graph1.nodes[node]['name'] == node_name1:
        node_number1 = node
        break
for node in Graph1.nodes():
    if Graph1.nodes[node]['name'] == node_name2:
        node_number2 = node
        break
for node in Graph1.nodes():
    if Graph1.nodes[node]['name'] == node_name3:
        node_number3 = node
        break
        
 # Utilisez Node2Vec
node2vec = Node2Vec(Graph4, dimensions=32, walk_length=30, num_walks=200, workers=4, p=1.0, q=0.1)

    # Apprenez les plongements
model = node2vec.fit(window=10, min_count=1)

    # Obtenez les vecteurs des nœuds
vectors = model.wv


edges_of_interest = []
edges_of_interest2 = []

for grap in Graph4.edges(data=True):
    source, target, data = grap
    relation = data.get("relation", None)

    if source == ressource1 and target != ressource1 and relation == bonneRelation:
        edges_of_interest.append((source, target, relation))

    if source == ressource2 and target != ressource2 and relation == bonneRelation:
        edges_of_interest2.append((source, target, relation))

# Supposons que votre ontologie utilise un espace de noms (namespace)
ontologie_namespace = Namespace("http://www.semanticweb.org/serge/ontologies/2022/10/pko-onto#")
df = pd.DataFrame(columns=['Entité A', 'Entité B', 'Similarité conceptuel', 'Similarité de plongement', 'Similarité Finale', 'Similarité de surface'])
   
for gra in edges_of_interest:
    for gra2 in edges_of_interest2:
        
        node_name1= gra[1]
        node_name2= gra2[1]
               
        for graphh in Graph3.edges:
            if graphh[1]== node_name1:
                concept1= graphh[0]
            if graphh[1]== node_name2:
                concept2= graphh[0]
    
        
        #print("#########################################")
        if node_name1 and node_name2 is not None:

            # Vous pouvez maintenant utiliser le numéro du nœud pour calculer la similarité avec d'autres nœuds.
            # Par exemple, pour mesurer la similarité cosinus entre le nœud recherché et un autre nœud (par exemple, 1) :
            similarity = vectors.similarity(node_name1, node_name2)
           
            s = consim(onto_file, concept1, concept2)
            
            finalSimilarity= 0.75*s+0.25*similarity
            
            #resourcetest1 = "ProjectManagementCourse"
            resourcetest1 = node_name1
            resourcetest2 = node_name2

                # Définissez l'URI de l'instance
            instance_uri1 = URIRef(f"{ontologie_namespace}{resourcetest1}")
            instance_uri2 = URIRef(f"{ontologie_namespace}{resourcetest2}")

                # Recherche de la triple avec "resourceName" dans l'espace de noms de l'ontologie
            resourceName_triple1 = (instance_uri1, ontologie_namespace.resourceName, None)
            resourceName_triple2 = (instance_uri2, ontologie_namespace.resourceName, None)
            
            resourceName_value1 = g.value(*resourceName_triple1)
            resourceName_value2 = g.value(*resourceName_triple2)
            #print(resourceName_value1)
            #print(resourceName_value2)
            similarity2 = name_similarity(resourceName_value1, resourceName_value2)
            
            ##seuil
           # if finalSimilarity <= 0.5 and finalSimilarity > 0.99:
           #     continue
            print(f"Similarity Concept between Node {concept1} and Node {concept2}: {s}")
            print(f"Similarity Embedding between Node {node_name1} and Node {node_name2}: {similarity}")
            print(f"Similarity Final between Node {node_name1} and Node {node_name2}: {finalSimilarity}")
            print("#########################################")
            print("#########################################")
            print("#########################################")
            new_row = pd.Series([node_name1, node_name2, s, similarity, finalSimilarity, similarity2], index=df.columns)
            warnings.filterwarnings("ignore", category=FutureWarning)
            df = df.append(new_row, ignore_index=True)
            
            
        else:
            print(f"Le nœud {node_name1} ou {node_name2} n'a pas été trouvé dans le graphe.")
        

for gra in edges_of_interest:
    for gra2 in edges_of_interest2:
        
        node_name1= gra[1]
        node_name2= gra2[1]
               
        for graphh in Graph3.edges:
            if graphh[1]== node_name1:
                concept1= graphh[0]
            if graphh[1]== node_name2:
                concept2= graphh[0]
    
        if node_name1 and node_name2 is not None:

            # Vous pouvez maintenant utiliser le numéro du nœud pour calculer la similarité avec d'autres nœuds.
            # Par exemple, pour mesurer la similarité cosinus entre le nœud recherché et un autre nœud (par exemple, 1) :
            similarity = vectors.similarity(node_name1, node_name2)
            #similarity = vectors.similarity(node_number1, node_number2)

            s = consim(onto_file, concept1, concept2)        
            finalSimilarity= 0.75*s+0.25*similarity
            
            #resourcetest1 = "ProjectManagementCourse"
            resourcetest1 = node_name1
            resourcetest2 = node_name2

                # Définissez l'URI de l'instance
            instance_uri1 = URIRef(f"{ontologie_namespace}{resourcetest1}")
            instance_uri2 = URIRef(f"{ontologie_namespace}{resourcetest2}")

                # Recherche de la triple avec "resourceName" dans l'espace de noms de l'ontologie
            resourceName_triple1 = (instance_uri1, ontologie_namespace.resourceName, None)
            resourceName_triple2 = (instance_uri2, ontologie_namespace.resourceName, None)
            
            resourceName_value1 = g.value(*resourceName_triple1)
            resourceName_value2 = g.value(*resourceName_triple2)
            #print(resourceName_value1)
            #print(resourceName_value2)
            similarity2 = name_similarity(resourceName_value1, resourceName_value2)
            

            if finalSimilarity >= 0.5 and finalSimilarity < 1.0:
                
                    print(f"Similarity Concept between Node {concept1} and Node {concept2}: {s}")
                    print(f"Similarity Embedding between Node {node_name1} and Node {node_name2}: {similarity}")
                    print(f"\033[91mSimilarity Final between Node {node_name1} and Node {node_name2}: {finalSimilarity}\033[0m")
    

                    print("#########################################")
                    print("#########################################")
                    
        else:
            print(f"Le nœud {node_name1} ou {node_name2} n'a pas été trouvé dans le graphe.")
            
for gra in edges_of_interest:
    for gra2 in edges_of_interest2:
        
        node_name1= gra[1]
        node_name2= gra2[1]
               
        for graphh in Graph3.edges:
            if graphh[1]== node_name1:
                concept1= graphh[0]
        #concept1= "HardResource"
        #concept2= "Objective"
        #concept3= "HardResource"
            if graphh[1]== node_name2:
                concept2= graphh[0]
    
        
        #print("#########################################")
        if node_name1 and node_name2 is not None:

            # Vous pouvez maintenant utiliser le numéro du nœud pour calculer la similarité avec d'autres nœuds.
            # Par exemple, pour mesurer la similarité cosinus entre le nœud recherché et un autre nœud (par exemple, 1) :
            similarity = vectors.similarity(node_name1, node_name2)
            #similarity = vectors.similarity(node_number1, node_number2)

            s = consim(onto_file, concept1, concept2)
            
            finalSimilarity= 0.75*s+0.25*similarity
            
            #resourcetest1 = "ProjectManagementCourse"
            resourcetest1 = node_name1
            resourcetest2 = node_name2

                # Définissez l'URI de l'instance
            instance_uri1 = URIRef(f"{ontologie_namespace}{resourcetest1}")
            instance_uri2 = URIRef(f"{ontologie_namespace}{resourcetest2}")

                # Recherche de la triple avec "resourceName" dans l'espace de noms de l'ontologie
            resourceName_triple1 = (instance_uri1, ontologie_namespace.resourceName, None)
            resourceName_triple2 = (instance_uri2, ontologie_namespace.resourceName, None)
            
            resourceName_value1 = g.value(*resourceName_triple1)
            resourceName_value2 = g.value(*resourceName_triple2)
            #print(resourceName_value1)
            #print(resourceName_value2)
            similarity2 = name_similarity(resourceName_value1, resourceName_value2)
            

            if finalSimilarity >= 0.5 and finalSimilarity < 1.0:
                if similarity2 >= 0.5 and similarity2 < 1.0: 
                    print(f"Similarity Concept between Node {concept1} and Node {concept2}: {s}")
                    print(f"Similarity Embedding between Node {node_name1} and Node {node_name2}: {similarity}")
                    print(f"Similarity Final between Node {node_name1} and Node {node_name2}: {finalSimilarity}")
               
               
                #resourcetest = node_name1
                #instance_uri = URIRef(f"{base_uri}#{resourcetest}")

                # Recherche de la triple avec rdfs:label
                #label_triple = (instance_uri, RDFS.label, None)
                #label_value = g.value(*label_triple)
                #print(label_value)
                
                
                      
                    # Texte en gras bleu
                    print(f"\033[1;34mSurface Similarity between {node_name1} and {node_name2}: {similarity2}\033[0m")

                    
                    print("\033[91m\033[1mFusion réussie avec succès\033[0m")

                    print("#########################################")
                    
                    instance_uri1 = URIRef(f"{ontologie_namespace}{node_name1}")
                    instance_uri2 = URIRef(f"{ontologie_namespace}{node_name2}")
                # Vérification si la triple existe et affichage du résultat
                
                    g.add((URIRef(instance_uri1), OWL.sameAs, URIRef(instance_uri2)))
                    #print(instance_uri1)
                    #print(instance_uri2)
                    g.serialize("result.ttl", format="turtle")
                    #t(g.serialize(format="turtle"))

        else:
            print(f"Le nœud {node_name1} ou {node_name2} n'a pas été trouvé dans le graphe.")
            
#PythonCourse
#PythonPractical
#usesResource
print(g)
```
